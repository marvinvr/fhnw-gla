b)

```python
B = []
for i in range(1, 7):
    tmp = []
    for j in range(1, 7):
        tmp.append(0 if i == j else i + j if i < j else i - j)
    B.append(tmp)
sp.Matrix(B)
```

```python
C = []
for i in range(1, 6):
    tmp = []
    for j in range(1, 6):
        tmp.append(0 if i + j < 6 else 1 if i + j == 6 else 2)
    C.append(tmp)
sp.Matrix(C)
```
