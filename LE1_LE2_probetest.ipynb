```python
import sympy as sp
```

```python
A = sp.Matrix([[1, 1, 1], [1, 1, 0], [1, 0, 0]])

a = sp.Symbol('a')
b = sp.Symbol('b')
c = sp.Symbol('c')

d = sp.Symbol('d')
e = sp.Symbol('e')
f = sp.Symbol('f')

g = sp.Symbol('g')
h = sp.Symbol('h')
i = sp.Symbol('i')

B = sp.Matrix([[a, b, c], [d, e, f], [g, h, i]])

AB = A * B
BA = B * A

gl = AB - BA

sol = sp.solve((gl[0,0], gl[0,1], gl[0,2], gl[1,0], gl[1,1], gl[1,2], gl[2,0], gl[2,1], gl[2,2]), a, b, c, d, e, f, g, h, i)

B = sp.Matrix([[sol[a], sol[b], sol[c]], [sol[d], sol[e], sol[f]], [g, h, i]])

print(B)

print(A * B)
print(B * A)

```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[-2], [-1], [0], [1], [2], [3]])
Y = np.array([[-1], [-1], [0], [0], [1], [1]])

m = np.hstack([X**5, X**4, X**3, X**2, X**1, X**0])
inv = np.linalg.inv(m)

v = np.dot(inv, Y)

x = np.arange(-2,3,0.01)

y = v[0]*x**5+v[1]*x**4+v[2]*x**3+v[3]*x**2+v[4]*x+v[5]*x**0

print(y)
```

```python
plt.plot(X, Y, 'r*', x, y, '-b')
```
