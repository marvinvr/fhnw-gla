## Aufgabe 4

```python
C = Matrix([
    [0, 5, 5, 5],
    [2, 3, 1, 5],
    [3, 2, -1, 6],
    [1, 0, -1, 1]
])
```

```python
def row_multiply(M, row, multiplier):
    M[row,:] = M[row,:] * multiplier
```

```python
def row_switch(M, i1, i2):
    row1 = M[i1,:].copy()
    M[i1,:] = M[i2,:].copy()
    M[i2,:] = row1
```

```python
def row_add(M, i1, i2, multiplier):
    M[i2,:] += M[i1,:] * multiplier
```

### b)

```python
row_switch(C, 0, 3)
```

```python
row_multiply(C, 1, 3)
row_add(C, 2, 1, -2)
row_add(C, 0, 2, -3)
```

```python
row_switch(C, 1, 2)
row_multiply(C, 1, 0.5)
row_add(C, 3, 2, -1)
row_add(C, 1, 3, -5)
```

### a)

```python
x = np.linspace( -5, 5, 100)
y1 = x - 1
y2 = -x - 1

plt.figure( figsize=(7,7) )
plt.plot( x, y1, label="y = x - 1")
plt.plot( x, y2, label="y = -x - 1")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(linestyle='--')
plt.gca().set_aspect('equal')
plt.legend()
```

```python
x = np.linspace(-10, 10, 100)

x1 = 0.5
y2 = x/3

plt.figure( figsize = (7,7) )
plt.axvline( x1, label="x = 0.5")
plt.plot(x, y2, label="y = x / 3", c="orange")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(linestyle='--')
plt.gca().set_aspect('equal')
plt.legend()
```

```python
np.linspace(-10, 10, 100)
y1 = 2.5*x - 2
y2 = -1 - 4/3*x

plt.figure( figsize=(7,7) )
plt.plot(y1, label="y = 2.5*x - 2")
plt.plot(y2, label="y = -1 - 4/3*x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(linestyle="--")
plt.gca().set_aspect("equal")
plt.legend()
```
