```python
import numpy as np
```

```python
A1 = np.array([[7, -2, -3], [-5, 1, 2], [-3, 1, 1]])
A2 = np.array([[2, 3, 4], [3, 4, 5], [4, 5, 6]])

b1 = np.array([[2], [0], [-1]])
b2 = np.array([[0], [0], [0]])
```

```python
_A1 = np.hstack([A1, b1])
print(np.linalg.matrix_rank(A1))
print(np.linalg.matrix_rank(_A1))
```

-> genau eine lösung

```python
_A2 = np.hstack([A2, b2])
print(np.linalg.matrix_rank(A2))
print(np.linalg.matrix_rank(_A2))
```

-> unendlich viele lösungen
